```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def func2(t, u):
    omega = math.radians(180*1.047198)
    phi = math.degrees(0)
    v = lambda vn: math.sqrt(vn[0]*vn[0] + vn[1]*vn[1] + vn[2]*vn[2])
    Ffric = lambda v: 0.0039 + (0.0058/(1 + math.e**(0.2*(v-35))))
    u0 = -Ffric(v(u))*v(u)*u[0] + 4.1*10**(-4)*omega*(u[2]*math.sin(phi) - u[1]*math.cos(phi))
    u1 = -Ffric(v(u))*v(u)*u[1] + 4.1*10**(-4)*omega*u[0]*math.cos(phi)
    u2 = -Ffric(v(u))*v(u)*u[2] - 4.1*10**(-4)*omega*u[0]*math.sin(phi) - 9.8
    u3 = u[0]
    u4 = u[1]
    u5 = u[2]
    
    return np.array([u0, u1, u2, u3, u4, u5])

iniciais = [38*math.cos(math.degrees(1)), 0, 38*math.sin(math.degrees(1)), 0, 0.0, 0]

intervalo = [0.0 , 0.5]

t = np.linspace(0.0, 0.5, 51)

u = solve_ivp(func2, intervalo, iniciais, t_eval=t).y 

plt.plot(t, u[0], label='X')
plt.plot(t, u[1], label='Y')
plt.plot(t, u[2], label='Z')
plt.legend()
plt.show()

plt.plot(t, u[3], label='X')
plt.plot(t, u[4], label='Y')
plt.plot(t, u[5], label='Z')
plt.legend()
plt.show()


```
